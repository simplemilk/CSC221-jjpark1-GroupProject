Import Library to Open URLs and use SSL.

```python
import ssl
import urllib
from bs4 import BeautifulSoup
```

Create the page variable in order for it to be used later.

Prettify the page as well to locate the table of interest.

```python
ssl._create_default_https_context = ssl._create_unverified_context

url = "https://en.wikipedia.org/wiki/South_Korea_national_football_team"
    
page = urllib.request.urlopen(url)

soup = BeautifulSoup(page, "lxml")

print(soup.prettify())
```

Find the title of the page.

```python
soup.title.string
```

Figure out the location of the table of interest.

```python
all_tables = soup.find_all("table")

print(f'Type of class for all_tables: {type(all_tables)}')

print(f'Length of all_tables: {len(all_tables)}')

for table in all_tables:
    
    print(f"TABLE: {table}\n")
```

Condense the tables down and try to find the table for Most Appearances.

```python
# Use find_all() to find all the <table...> tag with class_='wikitable sortable'
all_sortable_tables = soup.find_all("table", class_="wikitable sortable")

# Print the length of all_sortable_tables
print(f'length of all_sortable_tables: {len(all_sortable_tables)}')

# Iterate thru all_sortable_tables and print each table entry
for table in all_sortable_tables:
    
    print(f"SORTABLE TABLE: {table}\n")
```

The table of interest is the first table so splice that out of the previous code. 

```python
right_table = all_sortable_tables[0]

print(f"RIGHT TABLE: {right_table}\n")
```

Create empty lists to add the values into.

Classify the empty lists by their category while making a pandas DataFrame.

```python
A = []
B = []
C = []
D = []
E = []

for row in right_table.findAll("tr"):
    cells = row.findAll("td")
    if len(cells) == 5:
        A.append(cells[0].find(string=True).strip())
        B.append(cells[1].find(string=True).strip())
        C.append(cells[2].find(string=True).strip())
        D.append(cells[3].find(string=True).strip())
        E.append(cells[4].find(string=True).strip())
```

```python
# Import Pandas library 
import pandas as pd

# Create a DataFrame using column 'A' to start, then ...
df = pd.DataFrame(A, columns = ['Rank'])

# add each of the lists as a column in the DataFrame with the required column name
df['Player'] = B
df['Caps'] = C
df['Goals'] = D
df['Career'] = E


# Display the DataFrame
df
```

Convert the DataFrame into a CSV file

```python
df.to_csv('Group_Project.csv')

print('File Group_Project.csv created')
```

```python
# Creating a bar graph using the CSV data from Group_Project.csv
df = pd.read_csv('Group_Project.csv')

# Sorting the dataframe based on the 'Goals' column from highest to lowest goals scored
df_sorted = df.sort_values(by='Goals', ascending=True)

# Extracting the sorted players and goals
sorted_players = df_sorted['Player']
sorted_goals = df_sorted['Goals']

# Creating the bar graph
plt.figure(figsize=(20, 10))
bars = plt.barh(sorted_players, sorted_goals, color='dodgerblue')

# Adding actual Goals Scored by each player at the end of their bar
for bar in bars:
    plt.text(
        bar.get_width(),       
        bar.get_y() + bar.get_height() / 2,
        str(int(bar.get_width())), 
        va='center',             
        ha='left',               
        fontsize=18              
    )

# Naming the title, x-axis, and y-axis
plt.title('Most Goals Scored by South Korea National Football Team Players (1972-Present)', fontsize=28)
plt.ylabel('Player Names', fontsize=22)
plt.xlabel('Number of Goals', fontsize=22)
plt.yticks(fontsize=18)  
plt.xticks(fontsize=18)

plt.tight_layout()

# Displaying the plot
plt.show()
```
