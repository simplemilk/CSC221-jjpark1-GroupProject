Figure out the location of the table of interest.

```python
all_tables = soup.find_all("table")

print(f'Type of class for all_tables: {type(all_tables)}')

print(f'Length of all_tables: {len(all_tables)}')

for table in all_tables:
    
    print(f"TABLE: {table}\n")
```

Condense the tables down and try to find the table for Most Appearances.

```python
# Use find_all() to find all the <table...> tag with class_='wikitable sortable'
all_sortable_tables = soup.find_all("table", class_="wikitable sortable")

# Print the length of all_sortable_tables
print(f'length of all_sortable_tables: {len(all_sortable_tables)}')

# Iterate thru all_sortable_tables and print each table entry
for table in all_sortable_tables:
    
    print(f"SORTABLE TABLE: {table}\n")
```

The table of interest is the first table so splice that out of the previous code. 

```python
right_table = all_sortable_tables[0]

print(f"RIGHT TABLE: {right_table}\n")
```

Create empty lists to add the values into.

Classify the empty lists by their category while making a pandas DataFrame.

```python
A = []
B = []
C = []
D = []
E = []

for row in right_table.findAll("tr"):
    cells = row.findAll("td")
    if len(cells) == 5:
        A.append(cells[0].find(string=True).strip())
        B.append(cells[1].find(string=True).strip())
        C.append(cells[2].find(string=True).strip())
        D.append(cells[3].find(string=True).strip())
        E.append(cells[4].find(string=True).strip())
```

```python
# Import Pandas library 
import pandas as pd

# Create a DataFrame using column 'A' to start, then ...
df = pd.DataFrame(A, columns = ['Rank'])

# add each of the lists as a column in the DataFrame with the required column name
df['Player'] = B
df['Caps'] = C
df['Goals'] = D
df['Career'] = E


# Display the DataFrame
df
```

Convert the DataFrame into a CSV file

```python
df.to_csv('Group_Project.csv')

print('File Group_Project.csv created')
```
